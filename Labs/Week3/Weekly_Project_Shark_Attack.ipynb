# Top 10 countries in fatal attacks

Which countries had the most recorded deaths?
Let's start by grouping the information we are sorting by:

But first let's also remove the (Y/N) from Fatal so it's easier to manipulate.

```python
shark=shark.rename({'Fatal (Y/N)':'Fatal'}, axis=(1))
shark
```

```python
shark_country = shark.groupby(['Country', 'Fatal'])['Date'].count()
```

We then need to reset our index to remove all the 'N' variables:

```python
shark_country = shark_country.reset_index()
```

```python
shark_country
```

Let's filter by only lethal attacks

```python
shark_filter = shark_country['Fatal'] == 'Y'
shark_country_fatal = shark_country[shark_filter]
shark_country_fatal
```

Let's create a new df and change the header 'Date' to 'Total' so the information is clear

```python
shark_country_fatal=shark_country_fatal.rename({'Date':'Total'}, axis=(1))
shark_country_fatal
```

Let's organize by highest to lowest and see the top 10 countries with the most fatal attacks. 

```python
shark_country_fatal = shark_country_fatal.sort_values('Total',
                          ascending = False)

shark_country_fatal.head(10)
```

# Sorting by age and gender

Let's see what were the gender and ages of these victims

```python
shark_gender =shark.groupby(['Sex ','Age'])['Date'].count()

shark_gender

#shark_country = shark.groupby(['Country', 'Fatal'])['Date'].count()
#shark_country = shark.groupby(['Country', 'Fatal'])['Date'].count()
```

The data as is, needs to be cleaned before we can use it. Let's clear all non-numeric characters out of the 'Age' column

```python
def age(x):
    try: 
        return int(x)
    except:
        return np.nan
    '''
  Function searches the row for an int and returns it if found.
  If an int cannot be found, it replaces x with NaN
    
    '''
```

```python
shark_gender = shark_gender.reset_index()
```

```python
shark_gender['Age'] = shark_gender['Age'].apply(age)
```

```python
shark_gender
```

## How many people were underage?

```python
condition_age = (shark_gender['Age']<= 18)
condition_age.sum()
```
