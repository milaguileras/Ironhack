# Guessing Game

In this game, the user will attempt to guess the random number generated by the computer.
The user has 5 chances per round, with rounds getting harder as the game progresses.
On each round, chances will reset to 5. 


```python
import random
```

```python
number = random.randint(1,10)
```

```python
def phase1():
    chances = 0
    print('Input a number between 1-10: ')
    while chances < 5:
        guess = int(input())
        if guess == number:
            print('YAY!')
            phase2()
        elif guess < number:
            chances += 1
            if chances >= 5:
                print('Your guess is too low. You Lose. The correct number is:', number)
            else:
                print ('Your guess is too low. Guess higher than', guess)
        elif guess > number:
            chances +=1
            if chances >= 5:
                print('Your guess is too high. You Lose. The correct number is:', number)
            else:
                print ('Your guess is too high. Guess lower than', guess)    
     
```
